```python
dir_path='/content/drive/MyDrive/All fruits/temp1'
fruit_lt=os.listdir(dir_path)
label=[]
data=[]
for i in range(len(fruit_lt)):
  fruit_path=os.path.join(dir_path,fruit_lt[i])
  fruit_data=np.load(open(fruit_path,'rb'))
  imgs=550
  if len(fruit_data)<550:
    imgs=len(fruit_data)
  for j in range(imgs):
    label.append(i)
    data.append(fruit_data[j])
    #data[++k][:][:][:]=fruit_data[j]
    print(f'{i}-->{j}')
```
